 #####        using        parse_dates
 ###### for change data type to daytime64[ns] format
 In pandas, the parse_dates parameter is used to specify columns that should be parsed as dates when reading data from a file or when converting data into a DataFrame. 
 
 
ex:
 import pandas as pd
 
data = pd.read_csv('data.csv', parse_dates=['date_column', 'time_column'])
