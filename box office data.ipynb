```python
plt.figure(figsize=(6,11))
heat_map=df.corr()
sns.heatmap(heat_map,vmin=-1,vmax=1,annot=True)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot('rank','rating',data=df)
plt.title('rank & rating')
plt.show()
```

```python
sns.set_style('ticks')
sns.factorplot('rating','rank', hue='certificate',data=df)
plt.show()
```
